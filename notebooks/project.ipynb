### Problem 5 (20 points)

Build an ordinary least squares regression for the target using all the input columns.
Report the average loss of the model over the whole data set.
Plot the actual values vs the predicted outputs to compare them. 

```python
# YOUR CODE HERE

#Build the OLS 
X = df.drop(columns = 'target').values
y = df['target'].values

X_b = np.c_[np.ones((X.shape[0], 1)), X]
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

y_pred = X_b @ beta

#Find the average loss of this model
MSE = np.mean((y - y_pred) ** 2)
print(f"The average loss of this model: {MSE}")

#Plot the y and y_pred
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, color='green')
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red', label='Perfect Line')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.grid(True)



```

### Problem 8 (40 points)

Perform principal components analysis of the input columns.
Compute how much of the data variation is explained by the first half of the principal components.
Build a linear regression using coordinates computed from the first half of the principal components.

This problem depends on material from week 13.

```python
# YOUR CODE HERE
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Perform PCA of input columns
input_columns = [c for c in df.columns if c != 'target']
pca = PCA()
pca.fit(df[input_columns].dropna())
wine_components = pd.DataFrame(pca.components_, columns = input_columns)
wine_components



```

```python
# Compute how much of the data variation is explained by the first half of the principal components.
data = pca.explained_variance_
first_half = len(data) // 2
explained_first_half = np.sum(data[:first_half])
print(data)
explained_first_half
```

```python
# Build a linear regression using coordinates computed from the first half of the principal components.
pca_coordinates = pca.transform(df[input_columns].dropna())
half_coordinates = len(pca_coordinates) // 2
pca_X = pca_coordinates[:, :half_coordinates]
target = df.loc[df[input_columns].dropna().index, 'target']
model = LinearRegression()
model.fit(pca_X, target)
y_pred = model.predict(pca_X)
print(y_pred)
```

```python
#Compare the average training loss of this model to the previous model.
half_components = pca_X.shape[1] // 2
X_pca_half = pca_X[:, :half_components]

#Model 1: Linear Regression with PCA
model_1 = LinearRegression()
model_1.fit(X_pca_half, y)
y_pred_pca = model_1.predict(X_pca_half)
mse_pca = mean_squared_error(y, y_pred_pca)

# Model 2: Linear Regression with raw inputs
model_2 = LinearRegression()
model_2.fit(df[input_columns], y)
y_pred_raw = model_2.predict(df[input_columns])
mse_raw = mean_squared_error(y, y_pred_raw)

# Compare losses
print(f"MSE (PCA model): {mse_pca:.5f}")
print(f"MSE (Raw model): {mse_raw:.5f}")
print(f"Difference in training loss: {mse_pca - mse_raw:.5f}")
```

```python
# Plot actual targets vs predictions again. 
plt.figure(figsize=(8,6))
plt.subplot(1, 2, 1)
plt.scatter(y, y_pred_pca, alpha=0.6, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title("First Half PCA Model")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(y, y_pred_raw, alpha=0.6, color='green')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title("Raw Model")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.grid(True)
```

```python
# YOUR CODE HERE
from scipy import stats

# Calculate the z_score of the column 'pH'
z = np.abs(stats.zscore(df['pH']))

# Choose the threshold (=3)
threshold = 3

# Find the list of indexed rows has the outliers
outlier_indices = np.where(z > threshold)[0]
print(outlier_indices)
```

```python
outlier_array = z > 3
outliers = df[outlier_array]
# Find the row among outliers with the minimum pH value
min_ph_index = outliers['pH'].idxmin()
min_ph_value = df.loc[min_ph_index, 'pH']
plt.plot(df['pH'], marker='o', linestyle='', alpha=0.6, label='pH values')
plt.plot(min_ph_index, min_ph_value, 'ro', markersize=10, label=f'Outlier Row (index={min_ph_index})')
plt.title('Outlier Row Based on Minimum pH with Z-score > 3')
plt.xlabel('Row Index')
plt.ylabel('pH')
plt.legend()
```
